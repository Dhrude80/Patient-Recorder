```python
#Defining necesities
import json
Patients = {}
Options = {1: "Add Patient",
           2: "Remove Patient",
           3: "Edit Patient",
           4: "See Patient",
           5: "View all Patients",
           6: "Search patient",
           7: "Exit"
          }


while True:
    option = int(input("Enter your desired option."))

    if option == 1:
        ID = input("Enter the ID number of the patient you wish to add.")
        if ID in Patients:
            print("Patient already exists.")
        else:
            name = input("Enter the name of your patient.")
            try:
                age = int(input(f" What is the Age of {name} with ID number as {ID} ?"))
            except:
                print("Invalid Age")
                continue
            gender = input(f" What is the gender of {name} with ID number as {ID} ?")
            BG = input(f" What is the Blood Group of {name} with ID number as {ID} ?")
            disease = input(f" What disease does {name} with ID number {ID} have ?") 
            
            Patients[ID] = {
                "Name" : name,
                "Age" : age,
                "Gender" : gender,
                "Blood Group" : BG,
                "Disease" : disease
            }

    elif option == 2:
        ID = input("Enter the ID number of the patient you wish to remove.")
        if ID in Patients:
            del Patients[ID]
        else:
            print("Patient not in system.")
            continue

    elif option == 3:
        ID = input("Enter the ID number of the patient you wish to edit.")
        if ID in Patients:
            new_name = input("Enter the new name of your patient.")
            new_age = int(input(f" What is the Age of {new_name} with ID number as {ID} ?"))
            new_gender = input(f" What is the gender of {new_name} with ID number as {ID} ?")
            new_BG = input(f" What is the Blood Group of {new_name} with ID number as {ID} ?")
            new_disease = input(f" What disease does {new_name} with ID number {ID} have ?")

            Patients[ID] = {
                "Name" : new_name,
                "Age" : new_age,
                "Gender" : new_gender,
                "Blood Group" : new_BG,
                "Disease" : new_disease
            }
        else:
            print("Patient not found")
            continue

    elif option == 4:
        ID = input("Enter the ID number of the patient you wish to see.")
        if ID in Patients:
            print("Here is your the patient you wish to see :")
            print (Patients[ID])
        else:
            print("Patient not found")
            continue

    elif option == 5:
        for ID, details in Patients.items():
            print(ID, details["Name"])

    elif option == 6:
        ID = input("Enter the ID of the patient you are searching for.")
        if ID in Patients:
            print(f"Here are the details about Patient number {ID} :")
            print(Patients[ID])
        else:
            print("Patient not found")
            continue

    elif option == 7:
        confirm = input("Are you sure that you want to exit the program ?")
        if confirm.lower() in ["yes", "sure", "confirm", "ok"]:
            break
        elif confirm.lower() in ["no", "not quite", "probably not", "not sure"]:
            continue

            
#Storage parts
file = open("Patient_recorder.json","w")
json.dump(Patients, file)
file.close()
file = open("Patient_recorder.json","r")
Patients = json.load(file)
file.close()
```
